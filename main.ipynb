```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('kc_house_data.csv', sep=',')
df.head(10)
```

### Распределение стоимости

Продажи недвижимости отражают явную предпочтительность покупателей к низким ценам, приводя к высокой активности в этом сегменте. В среднем ценовом диапазоне наблюдается осторожность покупателей, а в высоком сегменте – минимальная активность, возможно, из-за ограниченного числа покупателей или специфических требований. Общий вывод: успешные сделки требуют тщательного анализа и адаптации стратегии к особенностям каждого ценового диапазона.

```python
plt.hist(df['price'], edgecolor='black')
plt.title("Распределение стоимости")
plt.xlabel("€", size = 17)
plt.ylabel("Кол-во домов", size = 12)
```

### Распределение квадратуры жилой площади
Анализ жилой недвижимости показывает, что основная часть предложения имеет площадь от 0 до 3000 квадратных метров, среди которых наиболее востребованы объекты от 1500 до 3000 квадратных метров. Небольшая доля приходится на жилье площадью от 3000 до 4000 квадратных метров, а очень крупные объекты свыше 4000 квадратных метров представлены минимально. Это подчеркивает необходимость адаптации стратегий продаж к разным сегментам площади для более успешных сделок.

```python
plt.hist(df['sqft_living'], edgecolor='black')
plt.title("Распределение квадратуры жилой площади")
plt.xlabel(r"$м^2$", size=15)
plt.ylabel("Кол-во домов", size=12)
```

### Распределение годов постройки
Анализ года постройки недвижимости выявляет, что наибольшая доля объектов (40%) возведена с 1940 по 1990 год, второй сегмент – 20% за период 1990-2020 год, и также 20% относится к постройкам с 1900 по 1940 год. Это свидетельствует о преобладании строительства в среднем и позднем XX веке, с небольшим вкладом от раннего XX века. Анализ годов поможет оптимизировать стратегии продаж и соответствовать запросам разнообразных покупателей.

```python
plt.hist(df['yr_built'], edgecolor='black', weights=np.ones(df.shape[0]) / df.shape[0])
plt.title("Распределение годов постройки")
plt.xlabel("Год", size = 12)
plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
plt.ylabel("Кол-во домов", size = 12)
```

### Распределение вида на набережную
Анализ указывает, что подавляющее большинство домов (99.2%) не располагают видом на набережную. На графике красные точки обозначают дома с набережной. Этот вывод подчеркивает ограниченность жилья с видом на водоем, что может повлиять на его привлекательность для потенциальных покупателей. Эффективные стратегии маркетинга могут быть направлены на преимущества других характеристик недвижимости, учитывая отсутствие набережной.

```python
data = df['waterfront'].value_counts()
values = data.values
plt.title("Распределение вида на набережную")
plt.pie(values, labels=['Есть','Нет'], autopct='%.1f%%');
plt.show()
waterfront_yes = df[df['waterfront']!=0]
plt.scatter(df['lat'], df['long'], c='blue', s=10, edgecolor = 'black')
plt.scatter(waterfront_yes['lat'], waterfront_yes['long'], c='red', s=9)
plt.gca().axes.get_xaxis().set_visible(False) 
plt.gca().axes.get_yaxis().set_visible(False) 
plt.show()
```

### Распределение этажей

Исходя из анализа графика, можно сделать вывод, что преобладающей частью на рынке недвижимости являются дома с 1 и 2 этажа. Объекты с 1.5 и 3 этажа представлены в меньшей степени, а дома с 2.5 и 3.5 этажа составляют крайне небольшую долю. Этот обзор подчеркивает предпочтение покупателей к более традиционным вариантам жилья с минимальным количеством этажей. Эффективные стратегии маркетинга и продаж должны учитывать эту динамику и акцентировать внимание на спросе на одно- и двухэтажные дома.

```python
plt.hist(df['floors'], edgecolor='black')
plt.title("Распределение этажей")
plt.xlabel("Этаж", size = 12)
plt.ylabel("Кол-во домов", size = 12)
```

### Распределение состояния домов
В оценке состояния недвижимости преобладают объекты с оценками 3 и 3.5, формируя значительную долю на рынке. Объекты с оценкой 4 составляют среднюю часть предложения. Оценки 4.5 и 5 присутствуют, но в небольших объемах, добавляя в рыночный ландшафт нюансы высокого статуса. Очень малое количество объектов оценено в диапазоне 1-2.5, что делает их редкими и не ликвидным предложением на рынке на фоне общей картины. Этот анализ подчеркивает необходимость учитывать динамику оценок в стратегиях продаж и маркетинге для успешного взаимодействия с разнообразными потребителями.

```python
plt.hist(df['condition'], edgecolor='black')
plt.title("Распределение состояния домов")
plt.xlabel("Оценка состояния", size = 12)
plt.ylabel("Кол-во домов", size = 12)
```

### Общее влияние характеристик
#### Влияние характеристик дома друг на друга 
Тепловая карта наглядно отображает взаимное влияние различных признаков недвижимости друг на друга. Яркие участки на карте указывают на тесные связи между определенными характеристиками объектов. Этот анализ позволяет четко выявить корреляции между различными аспектами недвижимости и определить, какие параметры взаимодействуют сильнее. Для более эффективного планирования и успешной продажи недвижимости важно учитывать эти взаимосвязи при разработке стратегий маркетинга и анализе спроса на рынке.
#### Влияние характеристик дома на его цену
Анализ тепловой карты выявляет, что наибольшее воздействие на цену недвижимости оказывают два ключевых фактора: жилая площадь и состояние дома. Яркие участки на карте, соответствующие этим признакам, подчеркивают их существенное влияние на ценовую динамику. Эти результаты подчеркивают важность данных параметров при определении стоимости недвижимости и позволяют эффективнее адаптировать стратегии маркетинга и продаж для успешного взаимодействия с требованиями рынка.

```python
df_new = df.drop(columns='date')
corr_matrix = df_new.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_price = corr_matrix.drop(columns=['id', 'bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long', 'sqft_living15', 'sqft_lot15'])
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')
plt.show()
sns.heatmap(corr_price, annot=True, linewidths=.5, cmap='coolwarm')
```
